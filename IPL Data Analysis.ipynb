```python
match_1.shape
```

```python
#to check how many fours ones sixes and all

srh['batsman_runs'].value_counts()
```

```python
# TOTAL SCORE CHECK

srh['batsman_runs'].sum()
```

```python
# FOR RCB

rcb=match_1[match_1['inning']==2]
rcb.shape
```

```python
rcb['batsman_runs'].sum()
```

```python
if srh['batsman_runs'].sum() > rcb['batsman_runs'].sum():
    print("SRH WON")
else :
    print("RCB")
```
